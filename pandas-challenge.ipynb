```python
import pandas as pd
```

```python
school_data_to_load = "/Users/alexisbloor/Desktop/pandas-challenge/schools_complete.csv"
student_data_to_load = "/Users/alexisbloor/Desktop/pandas-challenge/students_complete.csv"

school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

```

```python
school_data_complete_df = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete_df.head()
```

```python
# district summary
school_count = school_data['school_name'].count()
student_count = student_data['student_name'].count()
budget = school_data['budget'].sum()
avg_math = (student_data['math_score'].sum())/(student_count)
avg_read = (student_data['reading_score'].sum())/(student_count)
pass_math = student_data.loc[student_data['math_score'] >=70]
pct_math_pass = (len(pass_math)/student_count)*100

pass_read = student_data.loc[student_data['reading_score'] >=70]
pct_read_pass = (len(pass_read)/student_count)*100

overall = student_data.loc[(student_data['reading_score'] >=70) & (student_data['math_score'] >=70)]
overall_pct = (len(overall)/student_count)*100


summary_df = pd.DataFrame({"Total Schools": [school_count], 
                            "Total Students": [student_count], 
                            "Total Budget": [budget],
                            "Avg Math Score": [avg_math], 
                            "Avg Reading Score": [avg_read], 
                            "% Passing Math": [pct_math_pass], 
                            "% Passing Reading": [pct_read_pass],
                            "% Overall Passing": [overall_pct]})
summary_df

```

```python
#school summary

data_df = pd.merge(school_data, student_data, on="school_name")

school_index = school_data.set_index('school_name')
school_type = school_index['type']
school_size_students = school_index['size']
school_budget = school_index['budget']
budget_per_student = school_budget/school_size_students

index = data_df.set_index('school_name')
school = index.groupby(['school_name'])
avg_math_sch = school['math_score'].mean()
avg_reading_sch = school['reading_score'].mean()
total_students = school['student_name'].count()

math_sch = pass_math.groupby('school_name')
math_pass_sch = (math_sch['student_name'].count()/total_students)*100

read_sch = pass_read.groupby('school_name')
read_pass_sch = (read_sch['student_name'].count()/total_students)*100

overall_sch = (school_data_complete_df[(school_data_complete_df['math_score'] >= 70) & (school_data_complete_df['reading_score'] >= 70)].groupby('school_name')['student_name'].count()/total_students)*100

school_summary = pd.DataFrame({"School Type": school_type, 
                            "Total Students": school_size_students, 
                            "Total Budget": school_budget,
                            "Budget per Student": budget_per_student,
                            "Avg Math Score": avg_math_sch, 
                            "Avg Reading Score": avg_reading_sch, 
                            "% Passing Math": math_pass_sch, 
                            "% Passing Reading": read_pass_sch, 
                            "% Overall Passing": overall_sch})
school_summary = school_summary[["School Type", "Total Students", "Total Budget", "Budget per Student", "Avg Math Score", "Avg Reading Score", "% Passing Math", "% Passing Reading", "% Overall Passing"]]
school_summary
```

```python
# Sort and display the top five performing schools by % overall passing

top_five = school_summary.sort_values("% Overall Passing", ascending = False)
top_five.head(5)
```

```python
# Sort and display the five worst-performing schools by % overall passing.
bottom_five = school_summary.sort_values("% Overall Passing", ascending=True)
bottom_five.head(5)
```

```python
# math scores by grade
# Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school

nine_math = student_data.loc[school_data_complete_df['grade'] == '9th'].groupby('school_name')["math_score"].mean()
ten_math = student_data.loc[school_data_complete_df['grade'] == '10th'].groupby('school_name')["math_score"].mean()
eleven_math = student_data.loc[school_data_complete_df['grade'] == '11th'].groupby('school_name')["math_score"].mean()
twelve_math = student_data.loc[school_data_complete_df['grade'] == '12th'].groupby('school_name')["math_score"].mean()

grd_math_scores = pd.DataFrame({"9th": nine_math, "10th": ten_math, "11th": eleven_math, "12th": twelve_math})
grd_math_scores = grd_math_scores[['9th', '10th', '11th', '12th']]
grd_math_scores.round(3)
```

```python
# reading scores by grade
nine_read = student_data.loc[school_data_complete_df['grade'] == '9th'].groupby('school_name')["reading_score"].mean()
ten_read = student_data.loc[school_data_complete_df['grade'] == '10th'].groupby('school_name')["reading_score"].mean()
eleven_read = student_data.loc[school_data_complete_df['grade'] == '11th'].groupby('school_name')["reading_score"].mean()
twelve_read = student_data.loc[school_data_complete_df['grade'] == '12th'].groupby('school_name')["reading_score"].mean()

grd_read_scores = pd.DataFrame({"9th": nine_read, "10th": ten_read, "11th": eleven_read, "12th": twelve_read})
grd_read_scores = grd_read_scores[['9th', '10th', '11th', '12th']]
grd_read_scores.round(3)
```

```python
# scores by school spending
# Create a table that breaks down school performances based on average Spending Ranges (Per Student). 
# Use 4 reasonable bins to group school spending.

bins = [0, 585, 630, 645, 680]
spending_ranges = ['<$585', '$585-$630', '$630-$645', '$645-$680']
school_summary["spending ranges"] = pd.cut(school_summary["Budget per Student"], bins, labels=spending_ranges)


spend = school_summary.groupby("spending ranges")
spend = spend[["Avg Math Score", "Avg Reading Score", "% Passing Math", "% Passing Reading", "% Overall Passing"]]
spend.mean().round(2)
```

```python
# scores by school size

bins = [0, 1000, 2000, 5000]
school_range = ['<1000', '1000-2000', '2000-5000']
school_summary["school size"] = pd.cut(school_summary["Total Students"], bins, labels=school_range)


score_size = school_summary.groupby("school size")
score_size = score_size[["Avg Math Score", "Avg Reading Score", "% Passing Math", "% Passing Reading", "% Overall Passing"]]
score_size.mean().round(3)
```

```python
# scores by type
type_sum = school_summary.groupby("School Type")
type_sum = type_sum[["Avg Math Score", "Avg Reading Score", "% Passing Math", "% Passing Reading", "% Overall Passing"]]
type_sum.mean().round(3)
```
